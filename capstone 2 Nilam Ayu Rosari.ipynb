```python
plt.style.use('seaborn')
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Year_Birth', color='#01baef', linewidth=1.5, fliersize=4)
plt.title('Distribution of Year Birth', fontsize=16, fontweight='bold')
plt.xlabel('Year of Birth', fontsize=12)
plt.ylabel('Income', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

Terdapat 3 data outliers yang tahun lahirnya kurang dari 1900, sedangkan batas bawah dari Q1 yaitu sekitar tahun 1940, dimana jika asumsi data dikumpulkan adalah tahun 2016, maka umur pelanggan selain outliers masih dibawah 80 tahun. sehingga, dengan menghapus data outliers saja sudah dapat memenuhi asumsi dimana umur 80 tahun adalah batas umur produktif untuk berbelanja.

```python
def drop_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    if outliers.empty:
        print("Tidak ada data outliers dalam kolom", column)
    else:
        print("Data outliers dalam kolom", column, "adalah:")
        print(outliers.loc[:, [column]])
        df = df.drop(outliers.index)
        print("Data outliers telah dihapus.")
    
    return df
```

```python
df = drop_outliers(df, 'Year_Birth')
```

##### -) Grouping

```python
category_education = df['Education'].unique()
category_marital_status = df['Marital_Status'].unique()

print(f'Education : {category_education}, Marital Status : {category_marital_status}')
```

Dalam kategori pendidikan (education), terdapat beberapa kategori yang memiliki makna yang sama, seperti "2n cycle" dan "Master". Keduanya dapat dianggap sebagai kategori yang sama, sehingga akan digabung menjadi satu kategori.

Selain itu, dalam kategori status perkawinan (marital status), akan dilakukan penyederhanaan menjadi tiga kategori umum. Kategori "Absurd", "Alone", "Single", dan "YOLO" akan digabung menjadi kategori "Single". Kategori "Divorced" dan "Widow" akan digabung menjadi kategori "Divorced". Sedangkan kategori "Married" dan "Together" akan digabung menjadi kategori "Married".

Dengan melakukan penyederhanaan kategori ini, kita dapat mengurangi kompleksitas data dan membuatnya lebih mudah diinterpretasikan dan digunakan dalam analisis selanjutnya.

```python
def education_status(status):
    if status in ['Master', '2n Cycle']:
        return 'Master'
    else:
        return status

df['Education'] = df['Education'].apply(education_status)
```

```python
def transform_marital_status(status):
    if status in ['Absurd', 'Alone', 'Single', 'YOLO']:
        return 'Single'
    elif status in ['Divorced', 'Widow']:
        return 'Divorced'
    elif status in ['Married', 'Together']:
        return 'Married'
    else:
        return status

df['Marital_Status'] = df['Marital_Status'].apply(transform_marital_status)
```

### 2) Analisis Saluran Pemasaran Yang Paling Efektif.

```python
df['TotalBuyingChan'] = (df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases'])
top_100_customers = df.nlargest(100, 'TotalBuyingChan')
top_100_customers.head()
```

Untuk mengidentifikasi target pasar yang dapat menjadi fokus strategi pemasaran selanjutnya, saya telah memfilter data menjadi 100 member supermarket yang paling sering berbelanja baik dari via website, in-store ataupun katalog. Langkah ini bertujuan untuk memahami preferensi pembelian mereka, termasuk saluran pembelian yang paling sering digunakan dan tingkat frekuensi belanja yang mereka lakukan.
Dengan mengetahui preferensi pembelian dari anggota supermarket yang aktif, kita dapat memperoleh wawasan yang berharga dalam mengarahkan sumber daya pemasaran dengan lebih efektif. Pertama, kita dapat mengetahui saluran pembelian yang paling sering digunakan oleh anggota tersebut. Apakah mereka cenderung berbelanja secara langsung di toko fisik, ataukah mereka lebih memilih berbelanja secara online melalui platform e-commerce? Informasi ini akan membantu kita dalam menyesuaikan strategi pemasaran, baik itu dengan meningkatkan pengalaman belanja di toko fisik atau meningkatkan ketersediaan online melalui platform yang relevan.

```python
avg_web_purchases = top_100_customers['NumWebPurchases'].mean()
avg_catalog_purchases = top_100_customers['NumCatalogPurchases'].mean()
avg_store_purchases = top_100_customers['NumStorePurchases'].mean()

plt.figure(figsize=(10, 6))
colors = ['#C70039', '#FFC300', '#2E4053']
plt.bar(['Web', 'Catalog', 'Store'], [avg_web_purchases, avg_catalog_purchases, avg_store_purchases], color=colors, width=0.5)
plt.xlabel('Buying Access', fontsize=10, fontweight='bold')
plt.ylabel('Average Buying Access', fontsize=10, fontweight='bold')
plt.title('Buying Access Preference by Top 100 Members', fontsize=12, fontweight='bold')
max_value = max(avg_web_purchases, avg_catalog_purchases, avg_store_purchases)
plt.axhline(y=max_value, color='black', linestyle='--', linewidth=2)
plt.xticks(fontsize=9)
plt.yticks(fontsize=10)

plt.show()
```

Dari Bar plot diatas, dapat diambil kesimpulan bahwa saluran pembelian web memiliki kinerja yang tidak terlalu berbeda daripada catalog, hal ini bukan berarti penjualan melalui web nya tidak efisien hanya saja, pada tahun 2012-2014 belum terlalu marak dilakukan penjualan melalui online. Internet memang menjadi lebih umum untuk digunakan pada saat itu, namun belanja online belum mencapai tingkat popularitas yang sama seperti sekarang. Pada saat itu, masih ada sebagian konsumen yang lebih nyaman dengan belanja melalui katalog cetak atau mungkin tidak memiliki akses yang mudah ke internet.
Pengalaman belanja langsung ke store terlihat mendominasi karena memang dinilai lebih efisien sampai saat ini untuk produk-produk tertentu, hal ini disebabkan oleh beberapa hal : 
1. Pengalaman langsung: Toko fisik memberikan pengalaman langsung kepada konsumen. Mereka dapat melihat, meraba, dan mencoba produk secara langsung sebelum memutuskan untuk membeli. Ini memberikan kepercayaan dan kepuasan yang lebih tinggi kepada konsumen karena mereka dapat memastikan kualitas dan kesesuaian produk dengan kebutuhan mereka sebelum membeli.

Ini hal yang penting untuk produk-produk yang dibutuhkan dengan segera atau untuk konsumen yang tidak memiliki akses mudah ke pembelian online.

```python
grouped_age = top_100_customers['Age_Group']
grouped_income = top_100_customers['Income_Category']
grouped_education = top_100_customers['Education']
grouped_marital = top_100_customers['Marital_Status']

age_lastCamp = grouped_age.groupby(grouped_age).size()
income_lastCamp = grouped_income.groupby(grouped_income).size()
educ_lastCamp = grouped_education.groupby(grouped_education).size()
marital_lastCamp = grouped_marital.groupby(grouped_marital).size()

fig, axes = plt.subplots(2, 2, figsize=(16, 8))

age_max_value = age_lastCamp.max()
age_lastCamp.plot(kind='bar', ax=axes[0, 0], color=['#007200' if x == age_max_value else '#9ef01a' for x in age_lastCamp], width=0.3)
axes[0, 0].set_xlabel('Age Range', fontweight='bold')
axes[0, 0].set_ylabel('Response Count', fontweight='bold')
axes[0, 0].set_title('Age of Top 100 Membership', fontweight='bold')
axes[0, 0].grid(True, axis='y', linestyle='--', linewidth=0.5, color='lightgray', alpha=0.7)
axes[0, 0].set_xticklabels(age_lastCamp.index, rotation=0, ha='center')
axes[0, 0].axhline(age_max_value, color='black', linestyle='--')

wedgeprops = {'width': 0.2, 'edgecolor': 'white'}
income_max_value = income_lastCamp.max()
income_lastCamp.plot(kind='pie', ax=axes[0, 1], colors=['#db3a34' if x == income_max_value else '#cfdbd5' for x in income_lastCamp], autopct='%1.1f%%', startangle=140, wedgeprops=wedgeprops)
axes[0, 1].set_xlabel('')
axes[0, 1].set_ylabel('')
axes[0, 1].set_title('Income Category of Top 100 Membership', fontweight='bold')
axes[0, 1].axis('equal')

educ_max_value = educ_lastCamp.max()
educ_lastCamp.plot(kind='bar', ax=axes[1, 0], color=['#dc2f02' if x == educ_max_value else '#faa307' for x in educ_lastCamp], width=0.2)
axes[1, 0].set_xlabel('Educational Category', fontweight='bold')
axes[1, 0].set_ylabel('Response Count', fontweight='bold')
axes[1, 0].set_title('Educational Category of Top 100 Membership', fontweight='bold')
axes[1, 0].grid(True, axis='y', linestyle='--', linewidth=0.5, color='lightgray', alpha=0.7)
axes[1, 0].set_xticklabels(educ_lastCamp.index, rotation=0, ha='center')
axes[1, 0].axhline(educ_max_value, color='black', linestyle='--')

marital_max_value = marital_lastCamp.max()
marital_lastCamp.plot(kind='pie', ax=axes[1, 1], colors=['#db3a34' if x == marital_max_value else '#cfdbd5' for x in marital_lastCamp], autopct='%1.1f%%', startangle=25, wedgeprops=wedgeprops)
axes[1, 1].set_xlabel('')
axes[1, 1].set_ylabel('')
axes[1, 1].set_title('Marital Status of Top 100 Membership', fontweight='bold')
axes[1, 1].axis('equal')

fig.add_artist(plt.Line2D([0.56, 0.56], [0, 0.95], color='black', linewidth=1, linestyle='-'))
fig.suptitle('Segmentation of Top 100 Membership: Age, Income Category, Educational Background and Marital Status', fontsize=14, y=1, fontweight='bold')
plt.tight_layout()
plt.show()
```

##### Rekomendasi

Ada beberapa kegiatan yang bisa ditawarkan kepada segmen pasar yang sudah kita analisis sebelumnya, yaitu:
1.	Program Keanggotaan Eksklusif
Meluncurkan program keanggotaan eksklusif yang ditujukan untuk konsumen dengan tingkat pendapatan tinggi dan kebiasaan belanja yang konsumtif. Program ini memberikan berbagai manfaat eksklusif kepada anggota, seperti diskon khusus, akses ke produk-produk langka atau terbatas, undangan ke acara-acara eksklusif, dan yang paling utama adalah layanan pelanggan yang lebih baik. Memastikan staf supermarket ramah, membantu, dan responsif terhadap kebutuhan konsumen dapat membantu menciptakan pengalaman positif yang membuatnya ingin kembali berbelanja .
2.	 Penawaran Produk Premium
Menawarkan sejumlah produk premium yang tidak tersedia di supermarket lain kepada segmen konsumen tertentu. Misal, mereka dapat menawarkan anggur langka danberkualitas tinggi seperti Screaming Eagle atau daging berkualitas tinggi yang di impor langsung dari produsen terkemuka seperti Wagyu Kobe Beef Jepang yang terkenal dengan kelembutan dagingnya. Penawaran ini ditujukan kepada konsumen dengan preferensi dan kemampuan finansial untuk membeli produk-produk premium tersebut. Dengan menyediakan produk-produk eksklusif ini, supermarket dapat menarik perhatian konsumen yang lebih berkecimpung dalam pasar premium, yang akan mendatang margin keuntungan yang lebih tinggi pula.
3.	 Program Loyalitas Khusus
Mengembangkan program loyalitas khusus yang ditujukan untuk sekelompok konsumen yang dianggap memiliki potensi belanja yang tinggi, contohnya seperti Loyalty Program. Dengan menjadi anggota program loyalitas, supermarket dapat memberikan beberapa penawaran manfaat, seperti :
    1. Pengumpulan poin: Setiap kali pelanggan berbelanja dan menggunakan kartu keanggotaan atau aplikasi, mereka akan mendapatkan poin yang dapat dikumpulkan. Poin ini bisa ditukarkan dengan hadiah atau diskon khusus di masa mendatang.
    2. Diskon eksklusif: Anggota program loyalitas akan mendapatkan akses ke penawaran dan diskon khusus yang tidak tersedia untuk pelanggan biasa. Ini dapat membantu mereka menghemat uang saat berbelanja.
    3. Promosi khusus: juga dapat mengirimkan promosi khusus kepada anggota program loyalitas melalui email, pesan teks, atau notifikasi aplikasi. Ini memungkinkan anggota untuk mengetahui tentang penawaran terbaru, acara khusus, atau produk baru yang akan diluncurkan.

Namun, sebaiknya target pasar tidak hanya focus dalam menjual barang yang ekslusif dan ditujukan untuk konsumen yang memiliki jiwa konsumtif tinggi dan income yang memadai. Berikut adalah beberapa saran untuk menarik konsumen dengan minat belanja yang kurang:
1.	Membuat program loyalitas: Membuat program loyalitas yang sama seperti sebelumnya, hanya saja penawaran yang diberikan juga harus sesuai dengan kemampuan konsumen. Misal di applikasi loyalty terdapat level dari membership itu sendiri. Hal ini tentunya akan membantu supermarket dalam melakukan penawaran produk yang tepat. Konsumen yang merasa diperlakukan khusus tentunya akan memberikan respon positif sehingga meskipun mereka tidak mengalami perubahan dalam pembelanjaan, paling tidak mereka tidak berencana untuk mencari supermarket lain.
2.	Mengadakan acara atau promosi khusus: Mengadakan acara atau promosi khusus yang ditujukan untuk konsumen dengan minat belanja yang lebih rendah dapat menjadi cara untuk menarik perhatian mereka. Misalnya, mengadakan hari diskon khusus atau acara penjualan dengan harga terjangkau dapat menjadi daya tarik bagi konsumen dengan budget terbatas.
3. Penawaran paket atau bundel: Supermarket dapat menciptakan penawaran paket atau bundel yang menarik. Misalnya, Anda dapat membuat paket penawaran yang menggabungkan beberapa jenis wine dengan harga yang lebih terjangkau atau menyediakan paket daging yang berisi potongan-potongan kecil dengan harga yang lebih hemat.
4.	Menyediakan pelayanan yang baik: Pelayanan pelanggan yang baik sangat penting untuk menarik dan mempertahankan konsumen, Termasuk konsumen dengan minat belanja yang lebih rendah. Terkadang terdapat perbedaan treatment antara satu konsumen dengan konsumen yang lain, namun hal ini akan membawa kerugian terhadap supermarket. Meskipun konsumen tidak memperlihatkan pola belanja yang tinggi, hal itu tetap memberikan keuntungan untuk supermarket. Memiliki pelanggan yang banyak juga merupakan salah satu teknik marketing. Sehingga kehilangan satu konsumen akan sangat merugikan terutama jika alasannya adalah pelayanan yang buruk.
